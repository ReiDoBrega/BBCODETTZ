```
import urllib.request
import json
import re
import requests
import sys
import subprocess
import os
import time
from IPython.display import clear_output
from bs4 import BeautifulSoup

# Variables
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.8; rv:21.0) Gecko/20100101 Firefox/21.0'}
#@markdown ###INSIRA SUA KEY DO IMDB
api_key = '975b5d01' #@param {type:"string"}
#@markdown ---
#@markdown ###INSIRA O LINK/ID DO IMDB
IMDB_URL_OR_ID = '' #@param {type:"string"}
imdb_pattern = re.compile('(tt\d{7,8})')
youtube_pattern = re.compile('(https?://)?(www\.)?(youtube|youtu|youtube-nocookie)\.(com|be)/(watch\?v=|embed/|v/|.+\?v=)?(?P<id>[A-Za-z0-9\-=_]{11})')
imdb_id = re.findall(imdb_pattern, IMDB_URL_OR_ID)[0]

def get_imdb_template(imdb_id):
  imdb_url = f'http://www.omdbapi.com/?i={imdb_id}&apikey={api_key}&r=json&plot=full'
  #@markdown ---
  #@markdown ###INFO DO ARQUIVO/REMASTERIZADO/UPLOADER
  #@markdown (PREENCHA SEM ESPAÇOS, AO GERAR O BBCODE COLOQUE OS ESPAÇOS)
  Ano = "" #@param {type:"string"}
  Qualidade = "" #@param {type:"string"}
  Episodios = "" #@param {type:"string"}
  Idiomas = "" #@param {type:"string"}
  Legendas = "" #@param {type:"string"}
  Formato = "" #@param {type:"string"}
  Video = "" #@param {type:"string"}
  Resolucao = "" #@param {type:"string"}
  Audio = "" #@param {type:"string"}
  Duracao = "" #@param {type:"string"}
  Tamanho = "" #@param {type:"string"}
  Ripador = "" #@param {type:"string"}
  Remasterizador = "" #@param {type:"string"}
  Colaborador = "" #@param {type:"string"}
  Subber = "" #@param {type:"string"}
  Uploader = "" #@param {type:"string"}
  Server = "" #@param {type:"string"}
  #@markdown ---  
  #@markdown ###INFORMAÇÃO FILME/SERIE 
  Trailer = "" #@param {type:"string"}
  Screenshot_1 = "" #@param {type:"string"}
  Screenshot_2 = "" #@param {type:"string"}
  Screenshot_3 = "" #@param {type:"string"}
  Screenshot_4 = "" #@param {type:"string"}
  Agradecimentos = "" #@param {type:"string"}

  if Screenshot_1 != "":
    screenshot_1_list = Screenshot_1.split(" ")
    screenshot_1_enabled = True
  else:
    screenshot_1_enabled = False

  if Screenshot_2 != "":
    screenshot_2_list = Screenshot_2.split(" ")
    screenshot_2_enabled = True
  else:
    screenshot_2_enabled = False

  if Screenshot_3 != "":
    screenshot_3_list = Screenshot_3.split(" ")
    screenshot_3_enabled = True
  else:
    screenshot_3_enabled = False

  if Screenshot_4 != "":
    screenshot_4_list = Screenshot_4.split(" ")
    screenshot_4_enabled = True
  else:
    screenshot_3_enabled = False

  if Agradecimentos != "":
    agradecimentos_list = Agradecimentos.split(" ")
    agradecimentos_enabled = True
  else:
    agradecimentos_enabled = False

  if Trailer != "":
    trailer_list = Trailer.split(" ")
    trailer_enabled = True
  else:
    trailer_enabled = False

  if Ano != "":
    ano_list = Ano.split(" ")
    ano_enabled = True
  else:
    ano_enabled = False

  if Qualidade != "":
    qualidade_list = Qualidade.split(" ")
    qualidade_enabled = True
  else:
    qualidade_enabled = False

  if Episodios != "":
    episodios_list = Episodios.split(" ")
    episodios_enabled = True
  else:
    episodios_enabled = False

  if Idiomas != "":
    idiomas_list = Idiomas.split(" ")
    idiomas_enabled = True
  else:
    idiomas_enabled = False

  if Legendas != "":
    legendas_list = Legendas.split(" ")
    legendas_enabled = True
  else:
    legendas_enabled = False

  if Formato != "":
    formato_list = Formato.split(" ")
    formato_enabled = True
  else:
    formato_enabled = False

  if Video != "":
    video_list = Video.split(" ")
    video_enabled = True
  else:
    video_enabled = False

  if Resolucao != "":
    resolucao_list = Resolucao.split(" ")
    resolucao_enabled = True
  else:
    resolucao_enabled = False

  if Audio != "":
    audio_list = Audio.split(" ")
    audio_enabled = True
  else:
    audio_enabled = False

  if Duracao != "":
    duracao_list = Duracao.split(" ")
    duracao_enabled = True
  else:
    duracao_enabled = False

  if Tamanho != "":
    tamanho_list = Tamanho.split(" ")
    tamanho_enabled = True
  else:
    tamanho_enabled = False

  if Ripador != "":
    ripador_list = Ripador.split(" ")
    ripador_enabled = True
  else:
    ripador_enabled = False

  if Remasterizador != "":
    remasterizador_list = Remasterizador.split(" ")
    remasterizador_enabled = True
  else:
    remasterizador_enabled = False

  if Colaborador != "":
    colaborador_list = Colaborador.split(" ")
    colaborador_enabled = True
  else:
    colaborador_enabled = False

  if Subber != "":
    subber_list = Subber.split(" ")
    subber_enabled = True
  else:
    subber_enabled = False

  if Uploader != "":
    uploader_list = Uploader.split(" ")
    uploader_enabled = True
  else:
    uploader_enabled = False

  if Server != "":
    server_list = Server.split(" ")
    server_enabled = True
  else:
    server_enabled = False

  Path = "" #@param {type:"string"}
  if Path == "":
    mediainfo_enabled = False
  else:
    mediainfo_enabled = True
    tmp = subprocess.Popen('mediainfo --Logfile=/root/.nfo "{}"'.format(Path), stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
    time.sleep(5) # Wait 5s for the nfo file to populate properly
    with open('/root/.nfo', 'r', encoding='utf-8') as nfo_file:
      nfo = nfo_file.readlines()
      del nfo[1]
      nfo[1] = "Complete name                            : {}\n".format(os.path.basename(Path))
      nfo = ''.join(nfo)

  Link = "" #@param {type:"string"}
  if Link == "":
    print("Você esqueceu de colocar o link para download , por favor verifique!")
    sys.exit()

  #@markdown ---
  #@markdown ###INFORMAÇÃO DO SERVIDOR
  #@markdown (NÃO PREENCHA AS 2 ULTIMAS OPÇÕES)
  Likes = 'googledrive' #@param ["googledrive", "onedrive", "mega", "brfiles", "4shared", "1fichier", "zippyshare"]
  Posts = None #@param {type:"raw"}
  Thanks = None #@param {type:"raw"}

  def get_story_line(imdb_id):
    url = 'https://www.imdb.com/title/{}'.format(imdb_id)
    r = requests.get(url, headers=headers)
    data = r.text
    soup = BeautifulSoup(data, 'html.parser')
    story_line = soup.find_all('div', class_='inline canwrap')[0].find('span').text.lstrip()
    return story_line

  # Fetch data
  with urllib.request.urlopen(imdb_url) as imdb_url:
    imdb_data = json.loads(imdb_url.read().decode())

  # Process data
  text_dump = "[align=center][b][color=blue][size=220]{}[/size][/b][/color]\n".format(imdb_data['Title'])
  text_dump += "[b][size=190]{}[/size][/b]\n".format(imdb_data['Title'])
  text_dump += "[poster]{}[/poster][/align]\n".format(imdb_data['Poster'])
  text_dump += "[ttz]sinopse[/ttz]\n"
  text_dump += "{}\n".format(get_story_line(imdb_id))
  if ano_enabled:
    text_dump += "[ttz]informacoes[/ttz]\n"
    text_dump += "[b]Título Brasil / Original:[/b] {}\n".format(imdb_data['Title'],)
    text_dump += "[b]Gênero:[/b] {}\n".format(imdb_data['Genre'])
    for i in ano_list: 
      text_dump += "[b]Ano:[/b] {}\n".format(i)
    text_dump += "[b]Escrito Por:[/b] {}\n".format(imdb_data['Writer'])
    text_dump += "[b]Estrelando:[/b] {}\n".format(imdb_data['Actors'])
  if qualidade_enabled:
    for i in qualidade_list: 
      text_dump += "[b]Qualidade:[/b] {}\n".format(i)
  if episodios_enabled:
    for i in episodios_list: 
      text_dump += "[b]Episódios:[/b] {}\n".format(i)
  if idiomas_enabled:
    for i in idiomas_list: 
      text_dump += "[b]Idiomas:[/b] {}\n".format(i)
  if legendas_enabled:
    for i in legendas_list: 
      text_dump += "[b]Legendas:[/b] {}\n".format(i)
  if formato_enabled:
    for i in formato_list: 
      text_dump += "[b]Formato:[/b] {}\n".format(i)
  if video_enabled:
    for i in video_list: 
      text_dump += "[b]Video Info:[/b] {}\n".format(i)
  if resolucao_enabled:
    for i in resolucao_list: 
      text_dump += "[b]Resolução:[/b] {}\n".format(i)
  if audio_enabled:
    for i in audio_list: 
      text_dump += "[b]Áudio Info:[/b] {}\n".format(i)
  if duracao_enabled:
    for i in duracao_list: 
      text_dump += "[b]Duração por Episódio:[/b] {}\n".format(i)
  if tamanho_enabled:
    for i in tamanho_list: 
      text_dump += "[b]Tamanho por Episódio:[/b] {}\n".format(i)
  if ripador_enabled:
    for i in ripador_list: 
      text_dump += "[b]Ripador/Encoder: [/b] [cargo=rip] {}[/cargo]\n".format(i)
  if remasterizador_enabled:
    for i in remasterizador_list: 
      text_dump += "[b]Remasterizador: [/b] [cargo=rmz] {}[/cargo]\n".format(i)
  if colaborador_enabled:
    for i in colaborador_list: 
      text_dump += "[b]Colaborador(es):[/b] [cargo=colab] {}[/cargo]\n".format(i)
  if subber_enabled:
    for i in subber_list: 
      text_dump += "[b]Subber:[/b] [cargo=user] {}[/cargo]\n".format(i)
  if uploader_enabled:
    for i in uploader_list: 
      text_dump += "[b]Uploader:[/b] {}\n".format(i)
  if server_enabled:
    for i in server_list: 
      text_dump += "[b]Server:[/b] [b][color=blue] {}[/color][/b]\n".format(i)
  if trailer_enabled:
    for i in trailer_list:
      text_dump += "[ttz]trailer[/ttz]\n"
      text_dump += "[align=center][tube]{}[/tube]\n".format(i) 
  if mediainfo_enabled:
    text_dump += "[ttz][size=6][color=rgb(250, 197, 28)][b]Media Info[/b][/color][/size][/ttz]\n"
    text_dump += "[spoiler=MediaInfo][nfo]{}[/nfo][/spoiler]\n".format(nfo)
  text_dump += "[ttz]screenshots[/ttz]\n"
  if screenshot_1_enabled:
    for i in screenshot_1_list:
      text_dump += "[align=center][ss]{}[/ss]\n".format(i) 
  if screenshot_2_enabled:
    for i in screenshot_2_list:
      text_dump += "[ss]{}[/ss]\n".format(i) 
  if screenshot_3_enabled:
    for i in screenshot_3_list:
      text_dump += "[ss]{}[/ss]\n".format(i)
  if screenshot_4_enabled:
    for i in screenshot_4_list:
      text_dump += "[ss]{}[/ss][/align]\n".format(i)
  if agradecimentos_enabled:
    for i in agradecimentos_list:
      text_dump += "[ttz]agradecimentos[/ttz]\n"
      text_dump += "[b]{}[/b]\n".format(i) 
  text_dump += "[ttz]download[/ttz]\n"
  text_dump += "[hide][spoiler=Links][align=center][b][color=green]TITULO NO BRASIL[/color][/b][/align]\n"
  if Thanks:
    if Likes != 0 and Posts == 0:
      text_dump += "[center][HIDEREACT][HIDEREACTSCORE={}]{}[/HIDEREACTSCORE][/HIDEREACT][/center]".format(Likes, Link)
    elif Likes == 0 and Posts != 0:
      text_dump += "[center][HIDEREACT][HIDEPOSTS={}]{}[/HIDEPOSTS][/HIDEREACT][/center]".format(Posts, Link)
    elif Likes != 0 and Posts != 0:
      text_dump += "[center][HIDEREACT=1,2,3,4,5][HIDEREACTSCORE={}][HIDEPOSTS={}]{}[/HIDEPOSTS][/HIDEREACTSCORE][/HIDEREACT][/center]".format(Likes, Posts, Link)
    else:
      text_dump += "[center][HIDEREACT]{}[/HIDEREACT][/center]".format(Link)
  else:
    if Likes != 0 and Posts == 0:
      text_dump += "[center][HIDEREACTSCORE={}]{}[/HIDEREACTSCORE][/center]".format(Likes, Link)
    elif Likes == 0 and Posts != 0:
      text_dump += "[center][HIDEPOSTS={}]{}[/HIDEPOSTS][/center]".format(Posts, Link)
    elif Likes != 0 and Posts != 0:
      text_dump += "[servidor={}]{}[/servidor]".format(Likes, Link)
    else:
      text_dump += "[center]{}[/center]".format(Link)

  return text_dump

print(get_imdb_template(imdb_id))
```
